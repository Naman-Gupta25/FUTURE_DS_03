# Removing the unwanted columns 

**Student_ID** doesn't have any relation with any questions and there is no trend between them 

```python
df=df.drop(columns=['Student ID','Unnamed: 0'])
```

# Adding Overall_Satisfaction:



```python
df['Overall_Satisfaction']= df.mean(axis=1)
```

```python
df['Overall_Satisfaction'] =df['Overall_Satisfaction'].astype('int')
```

# Distribution Analysis 

```python
a=df.columns

for i in a :
    sns.displot(kind='hist',data=df,x=i,element='step',color='purple')
```

# observations :


1.Most children are not satisfied with the **Use of presentations**

2.Students are majorly 50-60% satisfied .

3.Majoly Students are satisfied with the **Stucture of the Course** .

# Bivariate Analysis :

# Questions:

1. when teacher is 'Well versed with the subject' then his way to 'Explains concepts in an understandable way' , relates how ?
2. Relation between 'Well versed with the subject' and 'Degree of difficulty of assignments'.
3. Relation between 'Solves doubts willingly' and 'Provides support for students going above and beyond'?


```python
sns.heatmap(pd.crosstab(df['Explains concepts in an understandable way'],df['Well versed with the subject'],normalize='columns'),annot=True)
```

```python
sns.heatmap(pd.crosstab(df['Degree of difficulty of assignments'],df['Well versed with the subject'],normalize='columns'),annot=True)
```

```python
sns.heatmap(pd.crosstab(df['Solves doubts willingly'],df['Provides support for students going above and beyond'],normalize='index'),annot=True)
```

# observations :


1.There is *no direct co-relation* between any respected parameters .

2.The way of *Explaining topics is not affiliated with teachers deep knowledge*.

3.Teachers well less well versed as well versed gives difficult Questions.

4.the *Least Willingly Doubt solving teachers provide  better support for students* but there is no overall co-relation.
